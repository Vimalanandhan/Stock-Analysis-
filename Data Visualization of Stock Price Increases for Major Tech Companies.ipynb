```python
import altair as alt
import pandas as pd
import numpy as np

# Sample data definition (replace this with your actual data)
data = {
    'Date': pd.date_range(start='2024-01-01', periods=3000, freq='D'),
    'AAPL': np.random.rand(3000) * 100,
    'GOOGL': np.random.rand(3000) * 100,
    'META': np.random.rand(3000) * 100,
    'MSFT': np.random.rand(3000) * 100
}

df1 = pd.DataFrame(data)

# Filter data
data_df = df1[0:3000]

# Reshape data for Altair
data_long = pd.melt(data_df, id_vars=['Date'], value_vars=['AAPL', 'GOOGL', 'META', 'MSFT'],
                    var_name='Symbol', value_name='Price')

# Convert 'Date' to datetime and extract month-year for aggregation
data_long['Date'] = pd.to_datetime(data_long['Date'])
data_long['MonthYear'] = data_long['Date'].dt.to_period('M').astype(str)

# Aggregate data by month and symbol
monthly_avg = data_long.groupby(['MonthYear', 'Symbol']).agg({'Price': 'mean'}).reset_index()

# Create Altair bar chart
chart = alt.Chart(monthly_avg).mark_bar().encode(
    x=alt.X('MonthYear:O', title='Month-Year', sort=None),
    y=alt.Y('Price:Q', title='Average Price'),
    color='Symbol:N',
    column='Symbol:N'  # Optional: Creates separate charts for each Symbol
).properties(
    width=alt.Step(20)  # Width of each bar
).interactive()

chart
```

```python
import altair as alt
import pandas as pd
import numpy as np

# Sample data definition (replace this with your actual data)
data = {
    'Date': pd.date_range(start='2024-01-01', periods=3000, freq='D'),
    'AAPL': np.random.rand(3000) * 100,
    'GOOGL': np.random.rand(3000) * 100,
    'META': np.random.rand(3000) * 100,
    'MSFT': np.random.rand(3000) * 100
}

df1 = pd.DataFrame(data)

# Filter data
data_df = df1[0:3000]

# Reshape data for Altair
data_long = pd.melt(data_df, id_vars=['Date'], value_vars=['AAPL', 'GOOGL', 'META', 'MSFT'],
                    var_name='Symbol', value_name='Price')

# Convert 'Date' to datetime
data_long['Date'] = pd.to_datetime(data_long['Date'])

# Aggregate data by symbol
avg_prices = data_long.groupby('Symbol').agg({'Price': 'mean'}).reset_index()

# Create Altair horizontal bar chart
chart = alt.Chart(avg_prices).mark_bar().encode(
    y=alt.Y('Symbol:N', sort='-x', title='Company'),
    x=alt.X('Price:Q', title='Average Price'),
    color='Symbol:N',
    tooltip=['Symbol:N', 'Price:Q']
).properties(
    title='Average Prices of Companies'
).interactive()

chart
```
